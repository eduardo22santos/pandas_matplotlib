```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cepea = pd.read_excel('CEPEA_20211104124240.xls')
```

```python
cepea.head()
```

```python
cepea.drop('À vista US$', axis=1,inplace=True)
```

```python
cepea[['dia','mes','ano']] = cepea['Data'].str.split('/', expand=True)
```

```python
cepea.head()
```

```python
cepea.drop('Data',axis=1,inplace=True)
```

```python
cepea.columns
```

```python
cepea2020 = cepea[['À vista R$', 'dia', 'mes', 'ano']].loc[cepea['ano']=='2020']
```

```python
cepea2020.tail()
```

```python
cepeaMedia = cepea2020.groupby('mes').mean()
```

```python
cepeaMedia.head(20)
```

```python
cepeaMedia.to_csv('milho_2020_media_de_precos.csv')
```
